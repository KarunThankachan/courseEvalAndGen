```python
f = open("OLIData.csv", "r")
lines = f.readlines()
f.close()

oli_edges = []
oli_concepts = set()
oli_graph = {}
for i in lines[1:]:
    line = i.strip()
    parts = line.split(",")
    oli_concepts.add(parts[0])
    oli_concepts.add(parts[1])
    if int(parts[2]) == 1:
        oli_edges.append([parts[0], parts[1]])
        if parts[0] not in oli_graph:
            oli_graph[parts[0]] = set()
        oli_graph[parts[0]].add(parts[1])
```

```python
three_size_edges = []
for i in sorted(oli_graph.keys()):
    for j in oli_graph[i]:
        if j in oli_graph.keys():
            for k in oli_graph[j]:
                three_size_edges.append([i, j, k])
```

```python
def reach(graph, c, c_dest, path, visited, ans):
    if c == c_dest:
        final_path = []
        done = {}
        for idx, i in enumerate(path):
            if i not in done:
                done[i] = idx
                final_path.append(i)
            else:
                final_path = final_path[:(done[i]+1)]
                done = {}
                for idx1, x in enumerate(final_path):
                    done[x] = idx1
        ans.append(final_path)
    if c in graph:
        for i in graph[c]:
            if c+","+i not in visited:
                visited[c+","+i] = True
                cur_list = path.copy()
                cur_list.append(i)
                reach(graph, i, c_dest, cur_list.copy(), visited, ans)
```
